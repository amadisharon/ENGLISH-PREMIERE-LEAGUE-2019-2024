```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

r = pd.read_csv('PL_23.csv')
#Get the Goal difference, win and loss probability
r['WPC'] = r['W'] / r['Pld']
r['LPC'] = r['L'] / r['Pld']
r['GD'] = r['GF'] - r['GA']

r
```

```python
a = pd.read_csv('PL_22.csv')

#Get the Goal difference, win and loss probability
a['WPC'] = a['W'] / a['Pld']
a['LPC'] = a['L'] / a['Pld']
a['GD'] = a['GF'] - a['GA']
a
```

```python
w = pd.read_csv('PL_21.csv')

#Get the Goal difference, win and loss probability
w['WPC'] = w['W'] / w['Pld']
w['LPC'] = w['L'] / w['Pld']
w['GD'] = w['GF'] - w['GA']
w
```

```python
plt.figure(figsize=(8,6))
sns.set_style('whitegrid')
SR= a.groupby('Team')['GD'].mean()
SR.plot(kind='bar')
plt.title('Teams Vs Goal Difference', size = 16)
plt.xlabel('Teams', color = 'red', size = 14)
plt.ylabel('Goal difference', color = 'green', size = 14)
plt.show
```

```python
plt.figure(figsize=(8,6))
sns.set_style('whitegrid')
SR= r.groupby('Team')['GD'].mean()
SR.plot(kind='bar')
plt.title('Teams Vs Goal Difference', size = 16)
plt.xlabel('Teams', color = 'red', size = 14)
plt.ylabel('Goal difference', color = 'green', size = 14)
plt.show
```
